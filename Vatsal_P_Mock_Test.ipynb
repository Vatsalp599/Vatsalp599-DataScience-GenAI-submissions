Now i informed AI to ensure and keep in its memory that Y value is loan status and rest are X values by saying " remember that Y value = loan status and rest are X values"

```python
y = df['loan_status']
X = df.drop('loan_status', axis=1)

print("Shape of X (features):")
print(X.shape)
print("Shape of y (target):")
print(y.shape)
print("\nFirst 5 rows of X:")
display(X.head())
print("\nFirst 5 values of y:")
display(y.head())
```

As one can see, the AI wrote code which gave output by segreggating the X values and Y value and giving first 5 row values of it

Now as i want to perform modelling by machine learning, i asked AI " divide dataset into training and testing split with 75% in training and 25% in testing and also stratify= Y value"

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

AI coded and separated the data into training and testing split in 75% training and 25% testing, it has given us the shape of X train, X test, Y train, Y test along with number of rows and columns with 37500 rows being in train and 12500 being in test which is exactly 75% and 25% of total rows in inital dataset

Before carrying out modelling with help of machine learning, i asked AI to scale the data with help of min-max scaling to ensure even weightage of all the X values, " scale the dataset with the help of min-max scaler

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Convert scaled arrays back to DataFrames for easier viewing and further processing (optional)
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns, index=X_test.index)

print("First 5 rows of scaled X_train:")
display(X_train_scaled.head())
print("\nFirst 5 rows of scaled X_test:")
display(X_test_scaled.head())
```

The code by AI has scaled the X train and X test dataset with the help of Min-Max scaler and i can see all the values are in between of 0 and 1 which is what we were expecting

Now i asked AI to perform logistic regression to predict whether loan will be approved or rejected and then get us scores of the model for analysis purpose, " here generate the logistic regression to predict if the loan is accepted or rejected and also give the scores of the model for analysis purpose"

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Initialize the Logistic Regression model
log_reg_model = LogisticRegression(random_state=42)

# Train the model on the scaled training data
log_reg_model.fit(X_train_scaled, y_train)

# Make predictions on the scaled test data
y_pred = log_reg_model.predict(X_test_scaled)

# Evaluate the Logistic Regression model's performance
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Logistic Regression Model Performance (Full Model):")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
```

The logistic regression model achieves an accuracy of 83.79%, indicating overall good performance. For the loan approved class, the precision is 84.46%, meaning that when the model predicts approval, it is correct about 84% of the time. The recall is 86.47%, showing that the model successfully identifies about 86% of all actual approved loans. This suggests the model is slightly better at capturing approved loans than avoiding false approvals, but loans for ineligible candidates getting not approved is the priority so we need less false positives and hence better precision score

For better precision, we need less false positives and model to have better score for precision, hence i asked AI to increase score of precision by using hyperparameters in the model and using gridsearch, " tune a model with help of hyperparameters and with help of grid search to have high precision and then run it"



```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'class_weight': [None, 'balanced']
}

print("Libraries imported and parameter grid defined.")
```

```python
logistic_model = LogisticRegression(solver='liblinear', random_state=42)
grid_search = GridSearchCV(estimator=logistic_model, param_grid=param_grid, scoring='precision', cv=5, verbose=1, n_jobs=-1)
grid_search.fit(X_train_scaled, y_train)

print(f"Best parameters: {grid_search.best_params_}")
print(f"Best precision score: {grid_search.best_score_}")
```

As we can see the AI coded for gridsearchCV, got best hyperparameter and tuned model with precision of 0.8594 and then ran the tuned model to get scores,
The logistic regression model achieves an overall accuracy of 83.61%. For loan approval predictions, the precision is 85.94%, meaning that when the model predicts approval, it is correct about 86% of the time. This high precision aligns with the objective of minimizing approvals for ineligible applicants.

Now i Asked AI to generate a model using decision trees" i want you to generate decision tree now to predict whether loan gets approved or rejected and give scores than for analysis purpose"

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Initialize the Decision Tree Classifier
decision_tree_model = DecisionTreeClassifier(random_state=42)

# Train the model on the scaled training data
decision_tree_model.fit(X_train_scaled, y_train)

# Make predictions on the scaled test data
y_pred_dt = decision_tree_model.predict(X_test_scaled)

# Evaluate the Decision Tree model's performance
accuracy_dt = accuracy_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test, y_pred_dt)
recall_dt = recall_score(y_test, y_pred_dt)
f1_dt = f1_score(y_test, y_pred_dt)

print(f"Decision Tree Model Performance:")
print(f"Accuracy: {accuracy_dt:.4f}")
print(f"Precision: {precision_dt:.4f}")
print(f"Recall: {recall_dt:.4f}")
print(f"F1-Score: {f1_dt:.4f}")
```

The Decision Tree model achieves an accuracy of 83.22%, indicating solid overall performance. For loan approval predictions, the precision is 84.83%, meaning that when the model predicts approval, it is correct about 85% of the time. Compared to logistic regression, this model offers similar accuracy but slightly lower precision, which may be relevant if minimizing false approvals is the priority

i asked AI to use hyperparameters to train decision tree model in order to increase precision score. "  i want you to train decision tree model with hyperparameters and tuning with gridsearchCV for the best precision score"

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [None, 5, 10, 20],
    'min_samples_split': [2, 10, 20],
    'min_samples_leaf': [1, 5, 10],
    'criterion': ['gini', 'entropy']
}

print("Libraries imported and parameter grid for Decision Tree defined.")
```

```python
dt_model = DecisionTreeClassifier(random_state=42)
grid_search_dt = GridSearchCV(estimator=dt_model, param_grid=param_grid, scoring='precision', cv=5, verbose=1, n_jobs=-1)
grid_search_dt.fit(X_train_scaled, y_train)

print(f"Best parameters for Decision Tree: {grid_search_dt.best_params_}")
print(f"Best precision score for Decision Tree: {grid_search_dt.best_score_}")
```

```python
best_dt_model = grid_search_dt.best_estimator_
y_pred_dt_tuned = best_dt_model.predict(X_test_scaled)

accuracy_dt_tuned = accuracy_score(y_test, y_pred_dt_tuned)
precision_dt_tuned = precision_score(y_test, y_pred_dt_tuned)
recall_dt_tuned = recall_score(y_test, y_pred_dt_tuned)
f1_dt_tuned = f1_score(y_test, y_pred_dt_tuned)

print("Tuned Decision Tree Model Performance:")
print(f"Accuracy: {accuracy_dt_tuned:.4f}")
print(f"Precision: {precision_dt_tuned:.4f}")
print(f"Recall: {recall_dt_tuned:.4f}")
print(f"F1-Score: {f1_dt_tuned:.4f}")
```

After hyperparameter tuning, the Decision Tree model shows improved performance with an accuracy of 85.37%. Precision is 86.99%, meaning that when the model predicts a loan approval, it is correct nearly 87% of the time, which aligns well with the goal of minimizing approvals for ineligible applicants. Recall is 86.32%, indicating that the model successfully identifies most eligible applicants. Compared to the previous version, tuning has significantly enhanced precision and overall performance, making this model more suitable for risk-sensitive loan approval decisions.
